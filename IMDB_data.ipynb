```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats
from sklearn.datasets import load_iris
import pandas as pd
```

```python
df_title_akas = pd.read_csv('zippedData/imdb.title.akas.csv.gz')
```

```python
df_title_basics = pd.read_csv('zippedData/imdb.title.basics.csv.gz')
```

```python
df_name_basics = pd.read_csv('zippedData/imdb.name.basics.csv.gz')
```

```python
df_title_crew = pd.read_csv('zippedData/imdb.title.crew.csv.gz')
```

```python
df_title_principals = pd.read_csv('zippedData/imdb.title.principals.csv.gz')
```

```python
df_title_ratings = pd.read_csv('zippedData/imdb.title.ratings.csv.gz')
```

```python
df_title_akas.info()
```

```python
df_merge1 = df_title_basics.merge(df_title_crew, how='inner', on='tconst')
```

```python
df_merge2 = df_merge1.merge(df_title_ratings, how='inner', on='tconst')
```

```python
df_merge3 = df_merge2.merge(df_title_principals, how='inner', on='tconst')
```

```python
df_imdb = df_merge3.merge(df_name_basics, how='inner', on='nconst')
```

```python
df_imdb_movies = df_imdb.drop_duplicates(subset='primary_title')
```

```python
subset = list(df_imdb_movies['genres'].value_counts().index[0:20])
```

```python
subset
```

```python
df_imdb_top_genres = df_imdb_movies[df_imdb_movies['genres'].isin(subset)]
```

```python
df_imdb_top_genres.groupby('genres').mean()['averagerating'].sort_values(ascending=False)
```

```python
df_imdb_directors = df_imdb[df_imdb['category'] == 'director']
```

```python
df_imdb_directors
```

```python
prolific_directors = list(df_imdb_directors['primary_name'].value_counts().index[0:500])
```

```python
df_imdb_directors = df_imdb_directors[df_imdb_directors['primary_name'].isin(prolific_directors)]
```

```python
df_imdb_directors.groupby('primary_name').mean()['averagerating'].sort_values(ascending=False)
```
